```python
from spellchecker import SpellChecker

spell = SpellChecker()
```

```python
def autocorrect_sentence(sentence):
    words = sentence.split()
    corrected_words = []

    for word in words:
        if word in spell:
            corrected_words.append(word)
        else:
            corrected_words.append(spell.correction(word))

    return " ".join(corrected_words)

# Test
input_text = input("\nEnter text ('exit' to stop): ")           # For instant use this -- "I lov pythn programing"
print("Original:", input_text)
print("Corrected:", autocorrect_sentence(input_text))
```

```python
def show_suggestions(word):
    if word in spell:
        return f"✔ '{word}' is correct."

    return f"Misspelled: {word}\nBest Guess: {spell.correction(word)}\nOther Suggestions: {spell.candidates(word)}"


word = input("Enter a word for which u wanna see the suggestion:")
show_suggestions(word)
```
